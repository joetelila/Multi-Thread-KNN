```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
font = {'size':13}
plt.rc('font', **font)
```

```python
df = pd.read_csv('optimized-knn-result.csv')
```

```python
df.head()
```

```python
#Plotting speed up: for all programs over the data.
data = ['xx_20k_s123', 'xx_50k_s456','xx_100k_s789']
implementations = ['STL PAR','STL PAR PinThread','OMP','FF']
speedups = ['stl_sp(nw)','stl_pin_sp(nw)','omp_sp(nw)','ff_sp(nw)']
scalability = ['stl_sc(nw)','stl_pin_sc(nw)','omp_sc(nw)','ff_sc(nw)']
efficiency = ['stl_eff(nw)','stl_pin_eff(nw)','omp_eff(nw)','ff_eff(nw)']
nw = ['1', '2', '4', '8', '16', '32', '64', '128', '256']
```

## Statistics for 20k

```python
df_20k = df[df['Test Data']==data[0]]
```

```python
# Plotting speedups
sp_list = []
for sp in speedups:
    sp_list.append(list(df_20k[sp]))
sc_list = []
for sc in scalability:
    sc_list.append(list(df_20k[sc]))
eff_lsit = []
for eff in efficiency:
    eff_lsit.append(list(df_20k[eff]))
```

```python
fig, axs = plt.subplots(3, 1, figsize=(13, 25))
for i in range(len(speedups)):
    axs[0].plot(nw, sp_list[i], label = implementations[i])
    axs[1].plot(nw, sc_list[i], label = implementations[i])
    axs[2].plot(nw, eff_lsit[i], label = implementations[i])
axs[0].legend()
axs[1].legend()
axs[2].legend()
axs[0].set_ylabel("Speedup")
axs[0].set_xlabel("Number of workers")
axs[1].set_ylabel("Scalability")
axs[1].set_xlabel("Number of workers")
axs[2].set_ylabel("Efficiency")
axs[2].set_xlabel("Number of workers")
```
